```python
#Declaracion de variables
ini=int(input('Dame el inicio de tu secuencia: '))
fin=int(input('Dame el fin de tu secuencia: '))
a="Fizz" 
b="Buzz"
#Generación de Algoritmo
for i in range(ini,fin+1):
  if (i%3==0 and i%5==0) :
    print(a+' '+b)
  elif i%3==0:
    print(a)
  elif i%5==0:
    print(b)
  else:
    print(i)
```
